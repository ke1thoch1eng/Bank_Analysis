DEFINING THE QUESTION

**The analytic puzzle that is to be solved**

**Metric of success**




1.   Determine insights into some of the key demographic factors that might drive individuals'financial outcomes.
2.   Provide indication of the state of financial inclusion in Kenya, Rwanda, Tanzania and Uganda.










**The context**

Traditionally, access to bank accounts has been regarded as an indicator of financial inclusion. Despite the proliferation of mobile money in Africa and the growth of innovative fintech solutions, banks still play a pivotal role in facilitating access to financial services. Access to bank accounts enables households to save and facilitate payments while also helping businesses build up their credit-worthiness and improve their access to other financial services. Therefore, access to bank accounts is an essential contributor to long-term economic growth.

DATA CLEANING

```python
# renaming  the columns with spelling errors

finance_df.rename(columns = {"The relathip with head": "relationship_with_head" , "Level of Educuation": "level_of_education"}, inplace=True)
finance_df.columns


```



---



```python
# ensuring all column names have the same format

finance_df.columns = finance_df.columns.str.lower().str.strip().str.replace(" ", "_")
finance_df
```

```python
# drop irrelevant fields
finance_df.drop(['uniqueid', 'year'], axis=1, inplace=True)
finance_df
```



---



```python

# checking for duplicates

finance_df.duplicated().sum()
```

```python
# checking for null values.

finance_df.isnull().sum()
```

```python
# drop null values

finance_df.dropna()
```

```python
# drop duplicates

df1 = finance_df.drop_duplicates()
df1
```

```python
# checking for outliers
 
df1.boxplot(figsize=(8, 6), fontsize=10)

# print size of the data with outliers present

print(df1.shape)

print("********************************************")

# removing outliers

Q1 = df1.quantile(0.25)
Q3 = df1.quantile(0.75)
IQR = Q3 - Q1
df2 = df1[~((df1 < (Q1 - 1.5 * IQR)) | (df1 > (Q3 + 1.5 * IQR))).any(axis=1)]

# print size of data after removal of outliers

print(df2.shape)
```

```python
# plot of the new data, without ouliers
df2.boxplot()
```

```python
# checking for anomalies
#
# calculating the lower and upper quantiles

q1_hse = df2["household_size"].quantile(0.25)
q3_hse = df2["household_size"].quantile(0.75)

# calculating inter-quartile range using values from above

iqr_hse = q3_hse - q1_hse

# alternative  fining quantiles using the percentiles from numpy library 

q1_h, q3_h = np.percentile(df2["household_size"], [25, 75]) 

# interqaurtile range

iqr_h = q3_h - q1_h

# comparison  if the two values are similar
print(iqr_hse, iqr_h)
```

Bivariate Analysis


```python
# checking the datatypes
df2.info()
```

```python
# creating a frequency table for the countries and accessibility to cell phone

countries_cellphone = pd.crosstab(index=df2['country'], columns=df2['cell_phone_access'])
print(countries_cellphone)

# plot chart for the frequencies represented in the crosstab dataframe

countries_cellphone.plot(kind="bar", figsize=(8,8),stacked=True)

# labelling axes and formatting chart

plt.title("Accessibility to cellphone by country")
plt.ylabel("Distribution across the population")
```

```python
# creating a frequency table dataframe

countries_house_size = pd.crosstab(index=df2['country'], columns=df2['household_size'])
print(countries_house_size)

# plot chart from the frequency table dataframe

countries_house_size.plot(kind="bar", figsize=(8,8),stacked=True)

# labelling axes and formatting chart

plt.title("Housesizes for the respondents by country")
plt.ylabel("Distribution across the population")
```

```python
# creating a frequency table for the countries and bank account possession

countries_bank_acc= pd.crosstab(index=df2['country'], columns=df2['has_a_bank_account'])
print(countries_bank_acc)

# plot chart for the frequencies represented in the crosstab dataframe 

countries_bank_acc.plot(kind="bar", figsize=(10,10),stacked=True)

# labelling axes and formatting chart

plt.title("Bank account possession by country")
plt.ylabel("Distribution across the population")
```

```python
# creating a frequency table dataframe

countries_gender = pd.crosstab(index=df2['country'], columns=df2['marital_status'])
print(countries_gender)

# plot chart from the dataframe

countries_gender.plot(kind="bar", figsize=(10,10))

# formatting and labelling chart

plt.title("Marital status of respondents by country")
plt.ylabel("Distribution across the population")
```

```python
# creating a frequncy table dataframe

gender_bank = pd.crosstab(index=df2['gender_of_respondent'], columns=df2['has_a_bank_account'])
print(gender_bank)

# plot of the relation of gender to individuals possession to bank account

gender_bank.plot(kind="bar", figsize=(10,10))

# formatting&labelling chart

plt.title("Relation of respondent age and possession of bank account")
plt.ylabel("Distribution across the population")
```

```python
# creating a frequency table  dataframe

house_size_bank = pd.crosstab(index=df2['household_size'], columns=df2['has_a_bank_account'])
print(house_size_bank)

# plot chart

house_size_bank.plot(kind="bar", figsize=(10,10), stacked=True)

# labelling chart

plt.title("Effect of household size to possession of bank account")
plt.ylabel("Distribution across the population")
```

Multivariate analysis.


```python
# Understanding the  relationship between country, level of education and bank account status
country_education = pd.crosstab(index=[df2['has_a_bank_account'],df2['country']], columns=df2['level_of_education'])

country_education.loc[('No', slice(None)),:].plot(kind='bar', figsize=(10,10))
plt.xticks([0,1,2,3], labels=['Kenya','Rwanda','Tanzania','Uganda'], rotation=0)
plt.xlabel('Education distribution per country')
plt.title('Respondents with no bank accounts')

country_education.loc[('Yes', slice(None)),:].plot(kind='bar', figsize=(10,10))
plt.xticks([0,1,2,3], labels=['Kenya','Rwanda','Tanzania','Uganda'], rotation=0)
plt.xlabel('Education distribution per country')
plt.title('Respondents with bank accounts')
```

The first observation is that the level of education with the highest number of respondents with no bank accounts is primary education.

In each country, the level of education with the highest number of respondents with bank accounts:

1. Kenya : Secondary Education  
2. Rwanda : Primary education  
1. Tanzania : Primary and Tertiary Education 
2. Uganda : Secondary education







```python
# Understanding the relationship between cell access, location and bank account status
loc_cell = pd.crosstab(index=[df2['has_a_bank_account'],df2['type_of_location']], columns=df2['cell_phone_access'])

loc_cell.loc[('No', slice(None)),:].plot(kind='bar', figsize=(12,10))
plt.xticks([0,1], labels=['Rural','Urban'], rotation=0)
plt.xlabel('Cell phone access distribution per location')
plt.title('Respondents with no bank accounts')
plt.legend(loc='upper right')

loc_cell.loc[('Yes', slice(None)),:].plot(kind='bar', figsize=(12,10))
plt.xticks([0,1], labels=['Rural','Urban'], rotation=0)
plt.xlabel('Cell phone access distribution per location')
plt.title('Respondents with bank accounts')
plt.legend(loc='upper right')
```

**Findings.**

 The following were observed:

*   A lion's share of the respondents in the four countries did not have or use a bank account.
*   A large number of the respondents are of age 30, while the mean of the ages of all respondents was 39.

*  Majority of the respondents have access to a cell phone.
*  Most  of the individuals in the population are married. 

*  A  large number of the respondent population were self-employed 
*  Majority of the respondents had only studied upto primary level of education; Very few had studied upto the tertiary level.
*  A large number of the respondents were heads of the household.
*  Majority of the respondents were female.

*  Majority of respondents in Kenya, Rwanda and Uganda are married while in Uganda, a huge population are single or never married
*  Kenya recorded a relatively high number of individuals possessing a bank account whereas Rwanda recorded a huge number of individuals without a bank account.


*  A large number of individuals that have access to a cell phone have a bank account. This could be due to accessiblity of mobile banking and online banking services all that be done via a cell phone.
*  The area of residence surprisingly had the opposite effect one expects  on bank account possession as more individuals living in the rural areas had a bank account compared to those in urban areas.
